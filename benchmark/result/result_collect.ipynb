```python
SIR = pd.read_csv('SIR_metric.csv')
UW = pd.read_csv('UW_metric.csv')
```

```python
results = pd.concat([SIR, UW]).iloc[:,1:]
```

```python


# results.to_csv('results.csv')
```

```python
train_sizes = [7,14,28]
test_sizes = [7,14]
methods = np.unique(results.method)
counties = np.unique(results.county)
results
```

```python




for train_size in train_sizes:
    for test_size in test_sizes:
        
        
        column_names = ['train','test','county']
        for method in methods:
            column_names.append(f'{method}_rel_err')
        table = pd.DataFrame()
        row = {}
        
        for county in counties:
            row['county']=county
            row['train']=train_size
            row['test']=test_size
            for method in methods:
                row[f'{method}_rel_err'] = round(results[(results.method==method)&(results.train_size==train_size)&(results.test_size==test_size)&(results.county==county)].rel_err.values[0],2)
            table = table.append(row, ignore_index=True)
        table.to_csv(f'train_{train_size}_test{test_size}.csv')
                
        
```
